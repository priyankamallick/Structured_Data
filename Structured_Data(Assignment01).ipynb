```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_parquet("/content/train.parquet")
test=pd.read_parquet("/content/test.parquet")
```

```python
train=train.dropna()
```

```python
train=train.drop_duplicates()
```

```python
train=train.drop("Date", axis=1)
```

```python
train.head()
```

```python
# One-Hot encode the Incident Columns
train = pd.get_dummies(train, columns=["Incident"])

# Group the incidents by Patient-Uid and sum the One-Hot encoded columns
train_group=train.groupby('Patient-Uid').sum()
train_group
```

```python
train_group.head()
```

```python
# Convert the Target Drug column to a Binary Column
train_group["Incident_TARGET DRUG"] = train_group["Incident_TARGET DRUG"].apply(lambda x: 0 if x == 0 else 1)
```

```python
train_group.dtypes
```

```python
test.head()
```

```python
test.describe()
```

```python
test=test.dropna()
```

```python
test=test.drop_duplicates()
```

```python
# A new column 'TARGET DRUG' is added to the DataFrame with default values of 0
test['TARGET DRUG'] = 0
test['DRUG_TYPE_18'] =0
```

```python
test.head()
```

```python
# One-Hot encode the Incident Columns
test = pd.get_dummies(test, columns=["Incident"])

# Group the incidents by Patient-Uid and sum the One-Hot encoded columns
test_group=test.groupby('Patient-Uid').sum()
test_group
```

```python
test_group.head()
```

```python
X=train_group.drop('Incident_TARGET DRUG',axis=1).values
y=train_group['Incident_TARGET DRUG'].values
```

```python
test_group.dtypes
```

# ML MODEL

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# split train data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score,f1_score
rf = RandomForestClassifier(n_estimators= 100, max_depth = 16, max_features='sqrt')
rf.fit(X_train, y_train)
predictions = rf.predict(X_val)
print("Training Accuracy :",rf.score(X_train,y_train))
print("Testing Accuracy :",rf.score(X_val,y_val))
print("F1 Score :", f1_score(y_val,predictions))
print("ROC AUC Score :", roc_auc_score(y_val,predictions))
```

```python
test_pred=test_group.drop("TARGET DRUG",axis=1).values
```

```python
predict= rf.predict(test_pred)
predict
```
